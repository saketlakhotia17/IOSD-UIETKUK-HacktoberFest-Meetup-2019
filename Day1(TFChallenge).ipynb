```
#using inline backend
%matplotlib inline


import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
#PLot Country-wise
df1 = df.groupby(['Country/Region']).mean()
df1_ = df1.iloc[2:,2:]
df1_['Mean'] = df1_.mean(axis = 1)
df1_ = df1_[df1_['Mean']>200]
df1_ = df1_.drop(['Mean'], axis = 1)
df1_1 = df1_.iloc[:,-1]
df1_1.plot(kind = 'bar', figsize = (15,7))
plt.title('COVID 19 cases as on 22 Mar 2020')
plt.show()
```

```

#Plot Date-wise
df2 = df1_.transpose()
df2 = df2.iloc[2:,1:]
```

```
df2.plot(figsize = (15,10))
plt.legend(loc = 'upper left')
plt.title('COVID 19 cases  by Country')
plt.show()

```

```

df3 = df2['India']
df3.plot(kind = 'bar', figsize = (15,5))
plt.title('COVID 19 cases - India')plt.show()
```

DATEWISE PLOT

```
dates=df.keys()
dates=dates[4:]
dates=list(dates)
```

```
cases_on_dates=[]
for i in dates:
  sum_of_cases=df[i].sum()
  cases_on_dates.append(sum_of_cases)
  print(f'{i}: {sum_of_cases}')
```

```
import random
```

```
import matplotlib.colors as mcolors
```

```
c=random.choices(list(mcolors.CSS4_COLORS.values()),k=len(dates))
plt.figure(figsize=(15,15))
plt.pie(cases_on_dates,colors=c)
plt.title('confirmed cases on basis of dates',size=30)
plt.legend(dates, loc='best')
plt.show()
```
